```python
import seaborn as sns

%matplotlib inline

tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

```python
tc = tips.corr()
```

```python
tc
```

```python
sns.heatmap(tc, annot=True, cmap='coolwarm',linecolor='green',linewidth=1)
```

```python
flights
```

```python
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
flights_pivot
```

```python
sns.heatmap(flights_pivot, cmap='magma', linecolor='green', linewidth=0.5)
```

```python
sns.clustermap(flights_pivot, standard_scale=1)
```
