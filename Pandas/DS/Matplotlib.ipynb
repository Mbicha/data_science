```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.plot(x,y)
```

```python
#Object Oriented Method
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,1,1])
axes.set_title('My Plot')
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.plot(x,y)
```

```python
fig.savefig('myplot.png')
```
